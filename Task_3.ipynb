```python
import pandas as pd
data = pd.read_csv("student_feedback.csv")
data.head()
```

```python
data.tail(10)
```

```python
data.columns
```

```python
data=data.drop(columns='Unnamed: 0',errors='ignore')
data
```

```python
data.info()
data.describe(include='all')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating= df.mean()
plt.figure(figsize=(10,5))
sns.barplot(x=avg_rating.values,y=avg_rating.index)
plt.title('avg feedback ratings')
plt.xlabel('values')
plt.ylabel('catogerie')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating= df.mean()
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.title('avg feedback ratings')
plt.xlabel('values')
plt.ylabel('catogerie')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
df=data.drop(columns='Student ID',errors='ignore')
avg_rating= df.mean()
plt.figure(figsize=(10,5))
sns.boxplot(data=df, orient='h' , palette='Set1')
plt.title('avg feedback ratings')
plt.xlabel('values')
plt.ylabel('catogerie')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='Degree of difficulty of assignments',bins=10)
plt.title('avg feedback ratings')
plt.xlabel('values')
plt.ylabel('catogerie')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.kdeplot(data=df['Course recommendation based on relevance'],fill=True)
plt.title('avg feedback ratings')
plt.xlabel('values')
plt.ylabel('catogerie')
plt.show()
```
